```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
# data of the bill in the restaurant
# the objective here is to predict the tip that the waiter will receive
```

```python
print(type(df))
print(df.shape)
```

```python
df.describe()
```

## Correlation heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection. This output is shown only for numerical variables.

```python
df.dtypes
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
# jointplot for analysing two features
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

## pairplot

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
# hue parameter is used to color the data points based on the value of another variable like sex

sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

## distplot or histogram

Dist plot helps us to check the distribution of the columns feature

```python
sns.histplot(df['tip'], kde=True,stat='density')
```

```python
sns.histplot(df['tip'], bins=10)
```

## Categorical plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about the following:

- boxplot
- violinplot
- countplot
- bar plot

```python
# count plot
sns.countplot(x='sex',data=df)
```

```python
# count plot
sns.countplot(x='day', data=df, palette='Set2')
```

## Barplot

the sames as count plot but we have to specify both x and y axises.


```python
sns.barplot(x='sex',y='total_bill', hue='smoker',data=df)
# play with variable names and parameters to get the desired output
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary (min, max, percentiles and outliers).


```python
sns.boxplot(x='day', y='total_bill', data=df,palette='rainbow',hue='sex')
```

```python
sns.boxplot(x='day', y='tip', data=df,palette='rainbow',hue='sex')
```

```python
# simple box plot, showing all numerical columns
sns.boxplot(df)
```

## Violin plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x='day', y='total_bill', data=df,palette='rainbow')
```
