# Matplotlib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
# with the above line, no need to write plt.show() every time
```

```python
import numpy as np
```

```python
# simple data
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# scatter plot
plt.scatter(x,y,c='g') # c is for color
plt.xlabel('X axix')
plt.ylabel('y axix')
plt.title('a simple 2 D graph')
# for saving the plot you can right click on it and save, or use the following to save it in your current wd
plt.savefig('simple 2D graph')
# because of %matplotlib inline, we don't need to write plt.show()
```

```python
# plot is another function inside matplotlib
y = x**2
#try the following options and plots one by one
# plt.plot(x,y)
# plt.plot(x,y,'r--')
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2, markersize=12)
```

```python
# creating subplots within one canvas
plt.subplot(2,2,1) # 2 rows, 2 columns, 1st plot
plt.plot(x,y,'b-')
plt.subplot(2,2,2) # 2 rows, 2 columns, 2nd plot
plt.plot(x,y)
plt.subplot(2,2,3) # 2 rows, 2 columns, 3rd plot
plt.plot(x,y,'r--')
plt.subplot(2,2,4) # 2 rows, 2 columns, 4th plot
plt.plot(x,y,'g*-',linestyle='dashed',linewidth=2, markersize=12)
```

```python
# another simple example
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('X axis caption')
plt.ylabel('Y axis caption')
plt.plot(x,y)
plt.show()
```

```python
# another simple example
x = np.arange(0,4*np.pi,0.1) # 0 to 4 pi with 0.1 steps
y = np.sin(x)
plt.title('sine wave form')
# plotting the curve
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,5*np.pi,0.1) # 0 to 5 pi with 0.1 steps
y_sin = np.sin(x)
y_cos = np.cos(x)

# set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2,1,1)

# make the first plot
plt.plot(x,y_sin,'b--')
plt.title('sine')

# set the second subplot as active, and make the second plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('cosine')

# show the figure.
plt.show()
```

```python
# bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show() # not neccessary because of %matplotlib inline
```

```python
# histogram for showing count or density of data
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=20) # default bins=10 and means the range of data is divided into 20 equal parts
plt.title('histogram')
plt.show()
```
